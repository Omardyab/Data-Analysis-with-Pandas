```python
import pandas as pd
import numpy as np
df_trip = pd.read_csv("./trip.csv",error_bad_lines=False)
df_station = pd.read_csv("./station.csv")
df_weather = pd.read_csv("./weather.csv")
df_trip.head(10)
df_station.head(10)
df_weather.head(10)
```

1-What is the average trip duration for a borrowed bicycle?

```python
avg_trip =df_trip['tripduration'].mean()
avg_trip
```

2-What’s the most common age of a bicycle-sharer?

```python
avg_age =2021-int(df_trip['birthyear'].mode())
avg_age
```

3-Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
df_weather['month'] = pd.DatetimeIndex(df_weather['Date']).month
average_prec=df_weather.groupby(['month']).mean()[['Precipitation_In']]
average_prec
```

```python
median_prec= df_weather.groupby(['month']).median()[['Precipitation_In']]
median_prec
```

4-What’s the average number of bikes at a given bike station?

```python
avg_num_bikes=df_station['install_dockcount'].mean()
avg_num_bikes
```
