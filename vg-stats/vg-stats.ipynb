```python
import pandas as pd
data=pd.read_csv('./vgsales.csv')
data
```

## 1- Which company is the most common video game publisher?

```python

most_common_publisher=data['Publisher'].mode()
most_common_publisher

# another way
# most_common_publisher =data['Publisher'].value_counts().idxmax()
# most_common_publisher
```

## 2-Whatâ€™s the most common platform?

```python
most_common_platform=data['Platform'].mode()
most_common_platform
# another way
# most_common =data['Platform'].value_counts().idxmax()
# most_common
```

## 3-What about the most common genre?

```python
most_common_genre=data['Genre'].mode()
most_common_genre
# another way
# most_common =data['Genre'].value_counts().idxmax()
# most_common
```

## 4- What are the top 20 highest grossing games?

```python
top_twenty=data[['Name','Global_Sales','Rank']].sort_values('Global_Sales',ascending=False).head(20)
top_twenty
```

## 5- For North American video game sales, whatâ€™s the median?

```python
med=data['NA_Sales'].median()
med
```

## 5.1- Provide a secondary output showing ten games surrounding the median sales output assume that games with same median value are sorted in descending order

```python
top_ten=data[data['NA_Sales'] == med].head(10)
top_ten
```

```python
top_ten=data[data['NA_Sales']== med][['Name']].head(10)
top_ten
```

## 5.2- assume that games with same median value are sorted in descending order

```python

med_desc=(data[data['NA_Sales']== med][['Name','Rank']].sort_values('Rank',ascending=False))[['Name']].head(10)
med_desc
```

## 6- For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
sd=(data['NA_Sales'].head(1)-data['NA_Sales'].mean())/data['NA_Sales'].std()
sd
```

## 7- The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
sales_compare=(data['Global_Sales'].mean(), data[data['Platform']=='Wii']['Global_Sales'].mean())
sales_compare
```
